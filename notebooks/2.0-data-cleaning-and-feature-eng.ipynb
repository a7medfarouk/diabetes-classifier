## Transofmation Logging

```python
class DataTransformationLog:
    """
    A comprehensive data cleaning logger that tracks all transformations
    """

    def __init__(self, dataset_name, analyst_name):
        self.dataset_name = dataset_name
        self.analyst_name = analyst_name
        self.start_time = datetime.now()
        self.log_entries = []
        self.summary_stats = {"initial": {}, "final": {}}

    def log_action(
        self,
        category,
        action,
        details,
        rows_affected=None,
        columns_affected=None,
        before_value=None,
        after_value=None,
    ):
        """
        Log a data cleaning action
        """
        entry = {
            "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            "category": category,
            "action": action,
            "details": details,
            "rows_affected": rows_affected,
            "columns_affected": columns_affected if columns_affected else [],
            "before_value": str(before_value) if before_value is not None else None,
            "after_value": str(after_value) if after_value is not None else None,
        }
        self.log_entries.append(entry)

    def to_dataframe(self):
        """Convert log to DataFrame"""
        return pd.DataFrame(self.log_entries)


# Initialize Log
log = DataTransformationLog(
    dataset_name="Employee_Attrition", analyst_name="Instructor"
)
print("Log Initialized.")
```

## Feature Scaling

```python
import numpy as np
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler

# Robust Scaling
robust_cols = ["HbA1c_level", "blood_glucose_level"]
robust_scaler = RobustScaler()
df_train_prediction[robust_cols] = robust_scaler.fit_transform(
    df_train_prediction[robust_cols]
)
log.log_action(
    category="Scaling",
    action="RobustScaler",
    details="Applied Robust Scaling to HbA1c_level and blood_glucose_level to handle outliers.",
    columns_affected=["HbA1c_level", "blood_glucose_level"],
)

# MinMax Scaling
mm = MinMaxScaler()
df_train_prediction[["age"]] = mm.fit_transform(df_train_prediction[["age"]])
log.log_action(
    category="Scaling",
    action="MinMaxScaler",
    details="Scaled age to [0, 1] range.",
    columns_affected=["age"],
)

# Log Transform + Standard Scaler for BMI (Prediction Dataset)
df_train_prediction["bmi"] = np.log1p(df_train_prediction["bmi"])
prediction_bmi_scaler = StandardScaler()
df_train_prediction[["bmi"]] = prediction_bmi_scaler.fit_transform(
    df_train_prediction[["bmi"]]
)
log.log_action(
    category="Scaling",
    action="Log Transform + StandardScaler",
    details="Applied Log Transform and StandardScaler to bmi",
    columns_affected=["bmi"],
)

# Log Transform + Standard Scaler for BMI (BRFSS Dataset)
df_train_brfss["BMI"] = np.log1p(df_train_brfss["BMI"])
brfss_bmi_scaler = StandardScaler()
df_train_brfss[["BMI"]] = brfss_bmi_scaler.fit_transform(df_train_brfss[["BMI"]])
log.log_action(
    category="Scaling",
    action="Log Transform + StandardScaler",
    details="Applied Log Transform and StandardScaler to BMI",
    columns_affected=["BMI"],
)
```

## Feature Discretization

```python
bins = [-1, 0, 13, 30]  # ranges deduced using domain knowledge
labels = ["Zero", "Moderate", "Severe"]


df_train_brfss["MentHlth_binned"] = pd.cut(
    df_train_brfss["MentHlth"], bins=bins, labels=labels
)
df_train_brfss["PhysHlth_binned"] = pd.cut(
    df_train_brfss["PhysHlth"], bins=bins, labels=labels
)


df_train_brfss = df_train_brfss.drop(columns=["MentHlth", "PhysHlth"])

display(df_train_brfss.head())
```

## Feature Encoding

```python
df_train_brfss = pd.get_dummies(
    df_train_brfss,
    columns=["MentHlth_binned", "PhysHlth_binned"],
    drop_first=True,
    dtype=int,
)
df_train_prediction = pd.get_dummies(
    df_train_prediction,
    columns=["gender", "smoking_history"],
    drop_first=True,
    dtype=int,
)
display(df_train_brfss.head())
display(df_train_prediction.head())
```

## Feature Interactions

```python
# Create the Cardio Score and drop the originals
df_train_brfss["Cardio_Comorbidity_Score"] = (
    df_train_brfss["HighBP"]
    + df_train_brfss["HighChol"]
    + df_train_brfss["Stroke"]
    + df_train_brfss["HeartDiseaseorAttack"]
)
df_train_brfss = df_train_brfss.drop(
    columns=["HighBP", "HighChol", "Stroke", "HeartDiseaseorAttack"]
)

# Create the Lifestyle Score and drop the originals
df_train_brfss["Lifestyle_Score"] = (
    df_train_brfss["PhysActivity"]
    + df_train_brfss["Fruits"]
    + df_train_brfss["Veggies"]
)
df_train_brfss = df_train_brfss.drop(columns=["PhysActivity", "Fruits", "Veggies"])

# Create the Cardio Risk Score and drop the originals
df_train_prediction["Cardio_Risk_Score"] = (
    df_train_prediction["hypertension"] + df_train_prediction["heart_disease"]
)
df_train_prediction = df_train_prediction.drop(
    columns=["hypertension", "heart_disease"]
)
```

```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

# BRFSS Dataset Feature Importance
rf_brfss = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf_brfss.fit(X_train_brfss, y_train_brfss)


fi_brfss_df = pd.DataFrame(
    {"Feature": X_train_brfss.columns, "Importance": rf_brfss.feature_importances_}
).sort_values(by="Importance", ascending=True)

# Plot
plt.figure(figsize=(10, 8))
plt.barh(fi_brfss_df["Feature"], fi_brfss_df["Importance"], color="skyblue")
plt.xlabel("Importance Score")
plt.title("Random Forest Feature Importance: BRFSS")
plt.tight_layout()
plt.show()

# Prediction Dataset Feature Importance
rf_pred = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf_pred.fit(X_train_prediction, y_train_prediction)


fi_pred_df = pd.DataFrame(
    {"Feature": X_train_prediction.columns, "Importance": rf_pred.feature_importances_}
).sort_values(by="Importance", ascending=True)

# Plot
plt.figure(figsize=(10, 6))
plt.barh(fi_pred_df["Feature"], fi_pred_df["Importance"], color="salmon")
plt.xlabel("Importance Score")
plt.title("Random Forest Feature Importance: Diabetes Prediction")
plt.tight_layout()
plt.show()
```

## Balancing Training Set (SMOTE)

```python
from imblearn.over_sampling import SMOTENC


print("BRFSS Dataset")
print("Before SMOTE:")
print(y_train_brfss.value_counts())

brfss_cat_indices = [i for i, col in enumerate(X_train_brfss.columns) if col != "BMI"]

smote = SMOTENC(categorical_features=brfss_cat_indices, random_state=42)
X_balanced_brfss, y_balanced_brfss = smote.fit_resample(X_train_brfss, y_train_brfss)

print("\nAfter SMOTE:")
print(y_balanced_brfss.value_counts())


print("\nDiabetes Prediction Dataset")
print("Before SMOTE:")
print(y_train_prediction.value_counts())

pred_num_cols = ["age", "bmi", "HbA1c_level", "blood_glucose_level"]
pred_cat_indices = [
    i for i, col in enumerate(X_train_prediction.columns) if col not in pred_num_cols
]

smote = SMOTENC(categorical_features=pred_cat_indices, random_state=42)
X_balanced_pred, y_balanced_pred = smote.fit_resample(
    X_train_prediction, y_train_prediction
)

print("\nAfter SMOTE:")
print(y_balanced_pred.value_counts())
```
