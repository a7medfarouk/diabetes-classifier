## Missing Values Analysis

```python

# Diabetes Prediction Dataset Missing Values Analysis
missing_report = pd.DataFrame({
'Feature': df_diabetes_prediction.columns,
'Missing_Count': df_diabetes_prediction.isnull().sum(),
'Missing_Percentage': (df_diabetes_prediction.isnull().sum() / len(df_diabetes_prediction) * 100),
'Data_Type': df_diabetes_prediction.dtypes })
# Sort by missing percentage
missing_report = missing_report.sort_values('Missing_Percentage', ascending=False)
# Flag columns with >20% missing
critical_missing = missing_report[missing_report['Missing_Percentage'] > 20]

display(missing_report.style.hide(axis="index"))
print(f"\nCritical: {len(critical_missing)} columns with >20% missing data")
```

## Data Validation Summary Report

```python
def create_validation_report(df, name):
    """Generate validation report for a dataset"""
    print(f"\n{'='*70}")
    print(f"VALIDATION REPORT: {name}")
    print(f"{'='*70}")
    
    # Basic metrics
    print(f"\nDATASET METRICS")
    print(f"Rows: {df.shape[0]:,}")
    print(f"Columns: {df.shape[1]}")
    print(f"Duplicates: {df.duplicated().sum()}")
    print(f"Missing Values: {df.isnull().sum().sum()}")
    
    # Data types
    print(f"\nDATA TYPES")
    for dtype, count in df.dtypes.value_counts().items():
        print(f"{dtype}: {count} columns")

# Generate reports for all datasets
create_validation_report(df_diabetes_prediction, "Diabetes Prediction Dataset")
create_validation_report(df_diabetes_brfss2015, "Diabetes BRFSS 2015")
create_validation_report(df_diabetes_brfss2021, "Diabetes BRFSS 2021")
```
